```python
# Welcome to your new notebook
# Type here in the cell editor to add code!
df = spark.read.table("qa_testing_raw")


```

```python
df.printSchema()
```

```python
display(df.limit(10))
```

```python
df = df.dropDuplicates()
```

```python
df = df.fillna({
    "QA_Notes_Raised": 0,
    "QA_Notes_Resolved": 0,
    "Bugs_Found": 0
})
```

```python
from pyspark.sql.functions import col

df = df.withColumn(
    "Test_Date",
    col("Test_Date").cast("date")
)
```

```python
from pyspark.sql.functions import when

df = df.withColumn(
    "QA_Resolution_Rate",
    when(
        col("QA_Notes_Raised") != 0,
        (col("QA_Notes_Resolved") / col("QA_Notes_Raised")) * 100
    ).otherwise(0)
)
```

```python
df.write.mode("overwrite").saveAsTable("qa_testing_clean")
```
